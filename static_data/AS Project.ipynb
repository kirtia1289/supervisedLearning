```python
# • Check for null values in the attributes.
df2 = pd.DataFrame(df1,columns=['Startup','Product','Funding', 'Event', 'Result', 'OperatingState'])
Dfn = df2.isnull()
print(Dfn)
```

```python
# 3. Data preprocessing & visualisation:
# Droping  null values.
fDf = df1.dropna()
fDf
```

```python
# Converting ‘Funding’ features to a numerical value.
fDf['Funding'] = fDf['Funding'].str.replace(',', '')
funding = fDf['Funding'].str.replace('$', '')
fund = list()
num_replace = {
    'K' : 1000,
    'M' : 1000000,
    'B' : 1000000000,
}

def pure_num(s):
    mult = 1.0
    while s[-1] in num_replace:
        mult *= num_replace[s[-1]]
        s = s[:-1]
    return float(str(s)) * mult

for x in funding: 
 fund.append(pure_num(x))
fund
```

```python
# • Plot box plot for funds in million.
div = 1000000
fundInMillion = list()

for y in fund:
    fundInMillion.append(y/div)  

df3 = fDf.copy()
#print(df2)
col = df3.columns.get_loc("Funding")
col
n = df3.columns[2]
# Drop that column
df3.drop(n, axis = 1, inplace = True)

df3[n] = fundInMillion
df3

sns.boxplot(x=df3['Funding'])
```

```python
# Calculating percentage of winners that are still operating and percentage of contestants that are still operating
rDf = fDf.copy()
rDf = rDf[rDf['OperatingState'].str.contains("Operating")]
wDf = rDf[rDf['Result'].str.contains("Winner")]
print(wDf)
cDf = rDf[rDf['Result'].str.contains("Contestant")]
print(cDf)

# As per the shape of Winner is 18 and final DF shape is 446 so % of winners that are still operating is
countFinal = fDf['Result'].count()
countWin = wDf['Result'].count()
perWin = countWin/countFinal*100
print("percentage of winners that are still operating:",perWin)

# As per the shape of Winner is 214 and final DF shape is 446 so % of winners that are still operating is
countContestant = cDf['Result'].count()
perCon = countContestant/countFinal*100
print("percentage of contestants that are still operating:",perCon)
```

```python
# • Write and perform your hypothesis along with significance test comparing the funds raised by companies across NY, SF and EU
# events from 2013 onwards.

# Null hypothesis: H0: Company which is Closed has more fund raised. Average fund raised by all companies are 17.24 M. This means 
# H0: μ = 17.24 M
# Alternate hypothesis: Company which is Operating has more fund raised.Average fund raised by all companies are 17.24 M. This means 
# H1: μ ≠ 17.24 M
cDf = df3.copy()
cDf = cDf[cDf['Event'].str.contains("NY|SF|EU")]
cDf = cDf[cDf['Event'].str.contains("2013|2014|2015|2016")]
print(cDf) 
comSum = cDf['Funding'].mean()
print("Average of Funds raised by NY,SF and EU",comSum)

# We can say null hypothesis can be rejected as the average funds raised by NY,SF and EU is higher then the average value.
```

```python
# To Plot the distribution plot comparing the 3 city events.
sns.displot(data=cDf, x="Event", hue="Event", kde = True)
```
